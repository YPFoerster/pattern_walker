```python
# from pattern_walker import fullProbPatternWalker as pw
from pattern_walker.utils import balanced_ditree,filter_nodes
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from pattern_walker.utils.plotting import walkerAnimation
```

```python

c=3 #number of Parts (coordination number of tree)
h=4 #height of tree (shortest distance between root and any leaf)
L=48 #number of bits to a pattern (length of glossary)
a_root=0.8 #expectation of root bits
Gamma_root=0.3 #noise on Act(root)-level 
tightness=0.8
Gamma=lambda tau:1-np.power(tau,1/(h-1)) #noise below Part-level
a_high=(1-a_root)*Gamma_root+a_root #expectation of Part-specific bits
a_low=(1-a_root)*Gamma_root+a_root/10 #expectation of Part-generic bits
Delta=1 #overlap between Parts
```

```python
# Make a rooted $c$-ary tree. 

tree,root = balanced_ditree(c,h)
```

```python
G=walkerAnimation(c,h,tree, root, L, a_root, a_low, a_high, Delta, Gamma(tightness), Gamma_root)
anim=G.anim()
plt.show()
```
